```python
import pandas as pd
from sqlalchemy import create_engine
import numpy as np
import csv
```

```python
mcd_facts = pd.read_csv('C:/Users/Khang/OneDrive/Desktop/openfoodfactsclean/mcd_menu_scrubbed.csv')
mcd_facts.head()
```

```python
only_coffee = mcd_facts.loc[mcd_facts["category"] == "Coffee & Tea", :]
only_coffee.head()
```

```python
split= only_coffee["product_name"]= only_coffee["product_name"].str.split(" ", n = 1, expand = True)
```

```python
split.head()
```

```python
only_coffee['type']=split[0]
only_coffee['description']=split[1]
only_coffee.head()
```

```python
only_coffee = only_coffee.drop(columns=['fat_g', 'carbohydrates_g', 'proteins_g'])
only_coffee.head()
```

```python
starb_facts = pd.read_csv('C:/Users/Khang/OneDrive/Desktop/openfoodfactsclean/star_menu_scrubbed.csv')
starb_facts = starb_facts.drop(columns=['fat_g', 'carbohydrates_g', 'proteins_g'])
starb_facts.head()
```

```python
starb_facts = starb_facts.rename(columns={"beverage_category": "id"})

starb_facts.drop_duplicates("id", inplace=True)
starb_facts.set_index("id", inplace=True)

starb_facts.head()
```

```python
only_coffee = only_coffee.rename(columns={"type": "id"})

only_coffee.drop_duplicates("id", inplace=True)
only_coffee.set_index("id", inplace=True)

only_coffee.head()
```

```python
merge_table = pd.merge(only_coffee, starb_facts, on="product_name")
merge_table
```
